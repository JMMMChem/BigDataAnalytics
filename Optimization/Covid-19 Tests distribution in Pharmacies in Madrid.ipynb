## General formulation

\begin{align*}
\underset{x_{i}}{\max} & \quad z=\sum_{i=1}^{n}c_{i}x_{i} = \sum_{i=1}^{n}(1/P_{i})x_{i} \\
\text{s.t.:}&\\
  &x_{i} \geq b_{i} \quad \forall i\\
  &x_{i} \leq d_{i} \quad \forall i\\
  &\sum_{i=1}^{n} x_{i} \leq 700000 \quad \forall i\\
  &x_{i} \geq 0 \quad \forall i 
\end{align*}

The variables and parameters in the formulae correspond to:

-$x_{i}$ is the number of tests in the suburb i

-$P_{i}$ is the population in the suburb i. 

-$F_{i}$ is the number of pharmacies in the suburb i. 

-$b_{i}$ is the minimum number of tests required at suburb i. 

-$d_{i}$ is the maximum test allocation capacity of pharmacies in suburb i.

```python
%%writefile assignment1.py

from pyomo.environ import *

model = AbstractModel()

#define the variables

model.n = Param(within=NonNegativeIntegers) #suburbs

#we define sets, that goes from 1 to the value of n
model.I = RangeSet(1,model.n) 

model.b = Param(model.I) #minimum number of test: b_j
model.d = Param(model.I) #maximum test allocation: d_i
model.F = Param(model.I) #number of pharmacies in a suburb: F_i
model.P = Param(model.I) #population in a suburb: P_i
model.c = Param(model.I) #inverse of population: 1/P_i
model.T = Param(model.I) #incidence of a suburb: I_i

# the next line declares a variable indexed by the set I
model.x = Var(model.I, domain=NonNegativeReals)

#definition of the objective function
def obj_expression(model): 
	return summation(model.c, model.x)


model.OBJ = Objective(rule=obj_expression, sense = maximize)

#minimum test constraint
def min_constraint_rule(model, i): # return the expression for the constraint for i
     return model.x[i] >= model.b[i]

# the next line creates one constraint for each member of the set model.I
model.min_Constraint = Constraint(model.I, rule=min_constraint_rule)

#maximum test allocation constraint
def max_constraint_rule(model, i): # return the expression for the constraint for i
     return model.x[i] <= model.d[i]
    
model.max_Constraint = Constraint(model.I, rule=max_constraint_rule)


#maximum tests constraint
def maxtest_constraint_rule(model): # return the expression for the constraint for i
     return sum(model.x[i] for i in model.I) <= 700000

model.maxtest_Constraint = Constraint(model.I, rule=maxtest_constraint_rule)   
```

##### c) Compute the sensitivities associated with each constraint, and interpret their values.

The sensitivity associated to each constraint shows how a change in the constraints affect the result of the objective function.

For example, given the constraint 

\begin{align*}
  &x_{i} \geq b_{i} \quad \forall i\\ 
\end{align*}

If there is a change in $b$, such that $b'=b+\delta b$, because of the linearity in the problem considered, there will also be a change in the solution $x'_{s}=x_{s}+\delta x_{s}$ and in $z'_{s}=z(x'_{s})=z_{s}+\delta z_{s}$

In the results above, it can be seen that the constraints with the least sensitivity are the ones related to the maximum allocation capacity in suburbs 49 (Valverde, in Fuencarral-El Pardo), 62 (Aluche, in Latina), 93 (Pueblo Nuevo, in Ciudad Lineal), 104 (Pinar del Rey, in Hortaleza) and 106 (Valdefuentes, in Hortaleza too). In every one of these districts from where the suburbs belong, the population density is among the latest ones in the city (in fact, Fuencarral-El Pardo is the one with the lowest density), so it makes sense that if we increase the maximum allocation capacity in the pharmacies in these suburbs, there will be no big change in the objective function. 

On the other hand, the constraints with the most sensitivity are the ones related to the maximum allocation capacity in suburbs are 13 (Atocha, in Arganzuela), 44 (El Pardo, in Fuencarral-El Pardo), 45 (Fuentelarreina, in Fuencarral-El Pardo), 57 (El Plantío, in Moncloa-Aravaca), 99 (Atalaya, in Ciudad Lineal), 117 (El Cañaveral, in Vicálvaro) and 127 (Aeropuerto, in Barajas). In this case, a district with a huge population density like Arganzuela appears to make sense to be included in this list, nevertheless, the others are among the ones with the least population density. However, they are among the suburbs with the biggest pharmacy-people rate, so that if a pharmacy increases its allocation capacity, it will be easier noticed in places where this rate is bigger.

Additionally, the minimal test constraints sensitivities are not shown, since the goal of the objective function is maximizing, therefore, the maximum capacity constraint has a much bigger effect on it than the minimal test. The objective function is somehow "pushing" the maximum capacity constraint to get to the optimal value.

##### d) Modify the problem in a) to impose some logical or conditional (linear) constraints that require the use of binary or integer variables. Implement and solve this new model  in Pyomo as an \AbstractModel" and interpret the results. Include the data used for the solution of the problem.


This new constraint can be based on the inclusion of the binary variable regarding whether a suburb should be restricted or not in order to maximize the restriction of areas where positive cases live. The maximum number of citizens to be restricted is 1/3 of the total population of the city. This political measure follows the strategy taken by the Comunidad de Madrid government, of an equilibrium between a health and an economic crysis (if other strategies were to be taken, like in Australia, the whole city should be restricted as a way to end up with the virus). In addition, no restrictions regarding maximum allocation capacity are considered since the interest does not have to do about the socioeconomic differencies between suburbs, but regarding absolute values of number of cases in the city.

$$
y_{i}=
\left\{\begin{array}{ll} 
1, & \text{if the suburb should be restricted,}\\
0, & \text{if it should not be restricted}
\end{array} \right.\quad i=1,2,3... n
$$

\begin{align*}

The relaxed problem, that is, the problem without integrality constraints, would look like:

  &\sum_{i=1}^{n} P_{i} y_{i} \leq 1/3\sum_{i=1}^{n} P_{i} \quad \forall i\\
  &0 \leq y_{i} \leq 1 \quad \forall i
\end{align*}

where $K_{i}$ is the number of cases in the suburb i. The closer $y_{i}$, the more advisable is the suburb i to be restricted, and the closer $y_{i}$, the less advisable is the suburb i to be restricted.

##### f) For the integer problem in d), solve several instances with different sizes (different numbers of variables) using randomly generated data. Comment on the impact of the size of the problem on the time required to solve it.

```python
#define the variables randomly created as said, with n as input parameter, the size of the problem, and n, P and T
#as output, our variables
def crear(n):
    P = np.random.randint(1000,70000, size=n);
    T = np.random.randint(200,500, size=n);
    return n,P,T
                    
```

```python
from pyomo.environ import *
opt = SolverFactory("glpk")
from pyomo.opt import SolverFactory
import numpy as np
import matplotlib.pyplot as plt
import time  

def trans_opt(n,P,T):
    model = AbstractModel()
        #model.n = Param(within=NonNegativeIntegers) #suburbs

#we define sets, that goes from 1 to the value of n
    model.I = RangeSet(1,n) 

   # model.P = Param(model.I) #population in a suburb: P_i
   # model.T = Param(model.I) #incidence of a suburb: I_i

# the next line declares a variable indexed by the set I
    model.y = Var(model.I, domain=Binary)

#definition of the objective function
    def obj_expression2(model): 
        return sum(T[i-1]*0.00001*P[i-1]*model.y[i] for i in model.I)


    model.OBJ = Objective(rule=obj_expression2, sense = maximize)

#maximum population constraint
    def maxpop_constraint_rule(model, i): # return the expression for the constraint for i
        return sum(P[i-1]*model.y[i] for i in model.I) <= 0.3333*sum(P[i-1] for i in model.I)

    model.maxpop_Constraint = Constraint(model.I, rule=maxpop_constraint_rule) 

#y constraint
    def y_constraint_rule(model, i): # return the expression for the constraint for i
        return model.y[i] <= 1

    model.y_Constraint = Constraint(model.I, rule=y_constraint_rule)

    instance = model.create_instance()
    results = opt.solve(instance)

    y_sol = np.zeros(n)
    
    for i in range(0, n):
        y_sol[i]=instance.y[i+1].value
       
    return instance.OBJ(), y_sol
      

```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

sizeN = np.arange(50,130,10)
timeTotal = np.zeros(len(sizeN))
res = np.zeros(len(sizeN))

for i in range(0,len(sizeN)):
    start = time.time()
    [n,P,T] = crear(sizeN[i]) #create the data
    [obj, y_sol] = trans_opt(n,P,T)
    res[i] = obj
    end = time.time()
    timeTotal[i] = end-start

plt.plot(sizeN,timeTotal)
plt.xlabel('Size [n]')
plt.ylabel('Time [s]')
plt.title('Size vs Time')
plt.show()
```

As expected, the solution of the problem gets harder and harder to compute as the size increases, since solving a problem with 50 observations is not the same as solving it with 130. The increasing gets steeper between 80 and 100, so that the difficulty increases quite a lot if size goes from 80 to 100 observations.
